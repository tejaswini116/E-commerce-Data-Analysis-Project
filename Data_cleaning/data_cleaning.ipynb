```python
orders=pd.read_csv("olist_orders_dataset.csv")
print(orders.columns)
print(orders.info())
orders.dropna(inplace=True)
orders.fillna(0)
orders["order_purchase_timestamp"]=pd.to_datetime(orders["order_purchase_timestamp"])
date_time=[
    "order_purchase_timestamp",
    "order_delivered_carrier_date",
    "order_delivered_customer_date",
    "order_estimated_delivery_date"
]
for col in date_time:
    orders[col]=pd.to_datetime(orders[col],errors="coerce")
orders.to_csv("orders_cleaned.csv",index=0)
```

```python
order_items=pd.read_csv("olist_order_items_dataset.csv")
order_items.dropna(inplace=True)
order_items.fillna(0)
print(order_items.columns)
print(order_items.info())
order_items["shipping_limit_date"]=pd.to_datetime(order_items["shipping_limit_date"])
order_items.to_csv("order_items_cleaned.csv",index=0)
```

```python
products=pd.read_csv("olist_products_dataset.csv")
products.dropna(inplace=True)
products.fillna(0)
print(products.columns)
print(products.info())
products.to_csv("products_cleaned.csv",index=0)
```

```python
sellers=pd.read_csv("olist_sellers_dataset.csv")
sellers.dropna(inplace=True)
sellers.fillna(0)
print(sellers.columns)
print(sellers.info())
sellers.to_csv("sellers_cleaned.csv")
```

```python
def cleaned_data(df):
    df.dropna(inplace=True)
    df.fillna(0,inplace=True)
    return(df)
```

```python
category_name=pd.read_csv("product_category_name_translation.csv")
order_payment=pd.read_csv("olist_order_payments_dataset.csv")
order_review=pd.read_csv("olist_order_reviews_dataset.csv")
geolocation=pd.read_csv("olist_geolocation_dataset.csv")
```

```python
cleaned_data(category_name)
print(category_name)
print(category_name.columns)
print(category_name.info())
category_name.to_csv("category_name_cleaned.csv",index=0)
```

```python
cleaned_data(order_payment)
print(order_payment)
print(order_payment.columns)
print(order_payment.info())
order_payment.to_csv("order_payment_cleaned.csv",index=0)
```

```python
cleaned_data(order_review)
print(order_review)
print(order_review.columns)
print(order_review.info())
order_review.to_csv("order_review_cleaned.csv",index=0)
```

```python
cleaned_data(geolocation)
print(geolocation)
print(geolocation.columns)
print(geolocation.info())
geolocation.to_csv("geolocation_cleaned.csv",index=0)
```
